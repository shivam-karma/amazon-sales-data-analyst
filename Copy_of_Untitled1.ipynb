Year-wise Sales Trend

Yearly Month-wise Sales Trend
Let's aggregate sales data by each month across different years and visualize the trend.

```python
  # Convert 'Order Date' column to datetime
Data['Order Date'] = pd.to_datetime(Data['Order Date'])
```

```python
# Extract year and month from 'Order Date' column
Data['Year'] = Data['Order Date'].dt.year
Data['Month'] = Data['Order Date'].dt.month
```

```python
# Monthly Sales Analysis
monthly_sales = Data.groupby(['Year', 'Month'])['Total Revenue'].sum().reset_index()

# Plot monthly sales trend
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_sales, x='Month', y='Total Revenue', hue='Year', marker='o', palette='viridis')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.legend(title='Year')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Yearly Sales Analysis
yearly_sales = Data.groupby('Year')['Total Revenue'].sum().reset_index()

# Plot yearly sales trend
plt.figure(figsize=(10, 6))
sns.barplot(data=yearly_sales, x='Year', y='Total Revenue', palette='mako')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Year-Monthly Sales Analysis
year_monthly_sales = Data.groupby(['Year', 'Month'])['Total Revenue'].sum().unstack().fillna(0)

# Plot year-monthly sales trend
plt.figure(figsize=(12, 8))
sns.heatmap(year_monthly_sales, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=0.5)
plt.title('Year-Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Year')
plt.tight_layout()
plt.show()
```

```python
#Total Sales:
total_sales = df['Total Revenue'].sum()
print("Total Sales:", total_sales)
```

```python
#Average Sales per Order:
average_sales_per_order = df['Total Revenue'].mean()
print("Average Sales per Order:", average_sales_per_order)
```

```python
#Top-selling Products
top_selling_products = df.groupby('Item Type').sum()['Total Revenue'].sort_values(ascending=False)
print("Top-selling Products:\n", top_selling_products)
```

Relationships:
****

```python
#Units Sold vs. Total Revenue:
correlation_units_revenue = df[['Units Sold', 'Total Revenue']].corr().iloc[0, 1]
print("Correlation between Units Sold and Total Revenue:", correlation_units_revenue)
```

```python
#Total Cost vs. Total Revenue:
correlation_cost_revenue = df[['Total Cost', 'Total Revenue']].corr().iloc[0, 1]
print("Correlation between Total Cost and Total Revenue:", correlation_cost_revenue)
```

```python
#Total Profit vs. Total Revenue:
correlation_profit_revenue = df[['Total Profit', 'Total Revenue']].corr().iloc[0, 1]
print("Correlation between Total Profit and Total Revenue:", correlation_profit_revenue)
```
